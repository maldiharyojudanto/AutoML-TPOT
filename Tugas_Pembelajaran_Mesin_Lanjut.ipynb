### **Drop data train**

```python
# drop/hapus kolom yang tidak diperlukan pada data train
df = df.drop(['id'], axis=1)
df.head()
```

```python
# mengecek data yang duplikat pada data train
df.duplicated().sum()
```

### **Transformasi data train**

```python
# melihat data yang terdapat pada kolom diagnosis
df['diagnosis'].value_counts()
```

```python
# mengubah data kategorik menjadi numerik
df['diagnosis'] = df['diagnosis'].map({"M":1,
                                       "B":0})
df.head()
```

```python
# menangani kolom yang bertipe data object (yang terdapat special character), selian kolom 'diagnosis' pada data train
df = df.replace('?', np.nan)
```

```python
# mengecek setiap kolom yang memiliki missing value (NaN)
df.columns[df.isnull().any()]
```

```python
# mengubah kolom bertipe data object ke float
df = df.astype({'concavity_mean':'float','concave points_mean':'float',
                'concavity_se':'float','concave points_se':'float',
                'concavity_worst':'float','concave points_worst':'float'})
```

```python
# menampilkan tipe data dari setiap kolom pada data train
df.info()
```

### **Visualisasi data train**

```python
df_var = ['perimeter_mean', 'area_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean',
          'radius_se', 'perimeter_se', 'radius_worst', 'perimeter_worst', 'area_worst', 'compactness_worst',
          'concavity_worst', 'concave points_worst']
```

```python
# menampilkan tipe data dari setiap kolom pada data test
df_test.info()
```

### **Drop data test**

```python
# drop/hapus kolom yang tidak diperlukan pada data test
df_test = df_test.drop(['id'], axis=1)
df_test.head()
```

```python
# mengecek data yang duplikat pada data test
df_test.duplicated().sum()
```

```python
# mengecek setiap kolom yang memiliki missing value (NaN)
df_test.columns[df_test.isnull().any()]
```

```python
df_test_var = ['perimeter_mean', 'area_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean',
          'radius_se', 'perimeter_se', 'radius_worst', 'perimeter_worst', 'area_worst', 'compactness_worst',
          'concavity_worst', 'concave points_worst']
```

## **Split data**

```python
# library sklearn untuk split data train dan test 
from sklearn.model_selection import train_test_split

x = df[df_var]  # semua row yang telah dipilih berdasarkan korelasinya dengan target
y = df['diagnosis']  # target

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# **Pemodelan**

## **Install TPOT**

## **Membangun model**

```python
# import library TPOTClassifier dan membangun model
from tpot import TPOTClassifier

model_tpot = TPOTClassifier(generations=5, 
                            population_size=10,
                            scoring='accuracy',
                            random_state=1,
                            verbosity=2, 
                            n_jobs=-1)
```

```python
# model fitting
model_tpot.fit(X_train, y_train)
```

```python
# import model terbaik yang berbentuk file .py ke google drive
model_tpot.export('/content/drive/MyDrive/Colab Notebooks/tpot_model_terbaik.py')
```

## **Evaluasi model**

```python
# import library confusion_matrix, precision score, recall
from sklearn.metrics import confusion_matrix, precision_score, recall_score
```

```python
# nilai confusion matrix
confusion_matrix(model_tpot.predict(X_test), y_test)
```

```python
# precision score
precision_score(model_tpot.predict(X_test), y_test)
```

```python
# recall score
recall_score(model_tpot.predict(X_test), y_test)
```

```python
# accuracy score
model_tpot.score(X_test, y_test)
```

## **Prediksi label dataset test**

```python
# menampung kelas/fitur df_test_var ke X_test_predict
X_test_predict = df_test[df_test_var]
```

```python
# melakukan prediksi terhadap dataset test
y_test_predict = model_tpot.predict(X_test_predict)
y_test_predict
```
